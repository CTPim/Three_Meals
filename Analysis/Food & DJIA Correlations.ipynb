```python
import pandas as pd
```

```python
# Import datasets as dataframes from csv files
corr_raw_df = pd.read_csv("C:/Users/jhillman/OneDrive/Desktop/Data Analytics Bootcamp/Three_Meals/Edited Data/Output/Correlation_data.csv")
corr_raw_df.head()
```

# Correlation of Percentages (1995-Present Day)


```python
col_list = list(corr_raw_df.columns)
col_list.remove("Beef_Pct_Change (Monthly)")
col_list.remove("Wheat_Pct_Change (Monthly)")
col_list.remove("Food - Pct_Change (Yearly)")
col_list.remove("Food - Pct_Change (Monthly)")
col_list.remove("Milk_Pct_Change (Monthly)")
col_list.remove("DJIA_Change_Percent")
col_list.remove("date_time")

percent_df = corr_raw_df.drop(col_list, axis=1)
percent_df = percent_df.set_index('date_time')
percent_df.head()
```

```python
# correlation between percentage changes
corr_percent_df = percent_df.corr()
corr_percent_df
```

# Correlation of Prices (1995-Present Day)


```python
col_list1 = list(corr_raw_df.columns)
col_list1.remove("Beef $/LB")
col_list1.remove("Wheat_Price_(USD/LB)")
col_list1.remove("Food - Index (1982-1984=100)")
col_list1.remove("Milk Cost per Gallon")
col_list1.remove("DJIA_Price")
col_list1.remove("date_time")

price_df = corr_raw_df.drop(col_list1, axis=1)
price_df = price_df.set_index('date_time')
price_df.head()
```

```python
# correlation between percentage changes
corr_price_df = price_df.corr()
corr_price_df
```

# Percentage Correlation (2008 - Present Day)

```python
percent_2008_df = percent_df.loc[percent_df.index >= "2008-01-01"]
percent_2008_df
```

```python
corr_percent_2008_df = percent_2008_df.corr()
corr_percent_2008_df
```

# Price Correlation (2008 - Present Day)

```python
price_2008_df = price_df.loc[price_df.index >= "2008-01-01"]
price_2008_df
```

```python
corr_price_2008_df = price_2008_df.corr()
corr_price_2008_df
```
