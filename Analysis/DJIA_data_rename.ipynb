```python
import pandas as pd
```

```python
# Import cleaned DJIA data and rename columns and re-clean dataset 
```

```python
file_path = "../Edited Data/cleaned_djia.csv"
djia_df = pd.read_csv(file_path)
djia_df
```

```python
# Drop old index
djia_df = djia_df.drop("Unnamed: 0", axis=1)
djia_df.head()
```

```python
# Change Date column name
djia_df = djia_df.rename(mapper={"Date": "date_time"}, axis=1)
djia_df.head()
```

```python
# Add DJIA_ to each column name to identify it
mapper = {}
col_list = list(djia_df.columns)
col_list.remove("date_time")

for i in col_list:
    mapper[i] = "DJIA_" + i
    
djia_df = djia_df.rename(mapper=mapper, axis=1)
djia_df.head()
```

```python
# Remove potential special characters from column names 
djia_df = djia_df.rename(mapper={"DJIA_Vol.": "DJIA_Volume", "DJIA_Change %": "DJIA_Change_Percent"}, axis=1)
djia_df
```

```python
# Replace old csv with this one and remove index
output_path = "../Edited Data/Output/cleaned_djia.csv"
djia_df.to_csv(output_path, index=False)
```
