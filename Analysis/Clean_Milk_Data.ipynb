```python
#Cleaning data for price of Milk

import pandas as pd
```

```python
# Load Average cost of milk (per gallon) data
milk_data = "C:/Users/jhillman/OneDrive/Desktop/Data Analytics Bootcamp/Three_Meals/Raw Data/Milk_Cost_1995 to 2023.csv"
milk_data_df = pd.read_csv(milk_data)
milk_data_df
```

```python
# Check datatypes
milk_data_df.dtypes
```

```python
#Chamge column names
milk_data_df = milk_data_df.rename(columns={'DATE':'Date','APU0000709112': 'Milk Cost per Gallon'})
milk_data_df
```

```python
#Change DATE to datetime
milk_data_df['Date'] = pd.to_datetime(milk_data_df["Date"], format="%Y-%m-%d")
milk_data_df
```

```python
# Check datatypes
milk_data_df.dtypes
```

```python
#Round price of milk
cleaned_milk_data_df = milk_data_df.round(decimals=3)
cleaned_milk_data_df
```

```python
# Check datatypes
milk_data_df.dtypes
```

```python
# Calculate and add pct_change column
cleaned_milk_data_df["Milk_Pct_Change (Monthly)"] = cleaned_milk_data_df["Milk Cost per Gallon"].pct_change(periods=1)*100
cleaned_milk_data_df.head(20)
```

```python
# Export as csv
cleaned_milk_data_df.to_csv('C:/Users/jhillman/OneDrive/Desktop/Data Analytics Bootcamp/Three_Meals/Edited Data/Output/cleaned_milk_data.csv')
```
