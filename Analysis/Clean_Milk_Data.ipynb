```python
#Cleaning data for price of Milk

import pandas as pd
```

```python
# Load Average cost of miilk (per gallon) data
milk_data = "Milk_Cost_1995 to 2023.csv"
milk_data_df = pd.read_csv(milk_data)
milk_data_df
```

```python
# Check datatypes
milk_data_df.dtypes
```

```python
#Chamge column names
milk_data_df = milk_data_df.rename(columns={'DATE':'Date','APU0000709112': 'Milk Cost per Gallon'})
milk_data_df
```

```python
#Change DATE to datetime
milk_data_df['Date'] = pd.to_datetime(milk_data_df["Date"], format="%Y-%m-%d")
milk_data_df
```

```python
# Check datatypes
milk_data_df.dtypes
```

```python
#Round price of milk
cleaned_milk_data_df = milk_data_df.style.format({
    'Milk Cost per Gallon': '{:,.2f}'.format,
})

cleaned_milk_data_df
```

```python
# Export as csv
milk_data_df.to_csv('cleaned_milk_data.csv')
```
