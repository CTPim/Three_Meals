```python
import pandas as pd
```

```python
file_path = "FRED_beef_cleaned.csv"
beef_df = pd.read_csv(file_path)
beef_df.head()
```

```python
rename_beef_df = beef_df.drop("Unnamed: 0", axis=1)
```

```python
rename_beef_df
```

```python
mapper = {"DATE": "date_time", "Price": "Beef $/LB"}

rename_beef_df = rename_beef_df.rename(mapper=mapper, axis=1)
rename_beef_df
```

```python
output_path = ("FRED_beef_cleaned2.csv")
rename_beef_df.to_csv(output_path, index=False)
```
